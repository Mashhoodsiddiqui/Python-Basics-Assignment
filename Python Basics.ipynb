#Theory Questions

**1. What is Python, and why is it popular?**
- Python is a general purpose programming language created by Guido Van Rossum. It is incredibly versatile flexible and beginner friendly, hence it is most popular programming language worldwide. Python is the most popular yet easiest programming language to learn.

**2. What is an interpreter in Python?**
- In Python, an interpreter is a program that reads and executes Python code line by line, translating it into machine readable intructions for the computer to understand and run.

**3. What are pre-defined keywords in Python?**
- Keywords in Python are predefined reserved words with special meaning so the interpreter can understand them.

**4.  Can keywords be used as variable names?**
- No, keywords cannot be used as variable names in most of programming languages because keywords have predefined meanings and are used to define the language's systax.

**5. What is mutability in Python?**
- An object is considered mutable if its state or value can be modified after it is created. This means that you can alter its internal data or attributes without creating a new object.

**6. Why are lists mutable, but tuples are immutable?**
- Lists are mutable because they are designed for dynamic data storage, allowing elements to be added, removed, or modified after creation. Tuples on the other hand, are immutable for efficieny and to ensure data integrity, once created their elements cannot be altered.

**7. What is the difference between “==” and “is” operators in Python?**
- The "==" and "is" operators in Python serve different comparision purposes-
  > Equality Operator (==) : It checks if the value of two operands are equal, it compares the content or data held by the objects.

  >Identity Operator (is) : It checks if two operands refer to the same object in memory, it compares the memory addresses of the objects.

**8. What are logical operators in Python?**
- Logical operators are used to combine or modify Boolean values (True or False). There are three main logical operators "and" "or" and "not".

**9. What is type casting in Python?**
- Type casting, also known as type conversion, is the process of changing a variable's data type into another data type. In Python this can be done using *implicitly* or *explicitly*.

**10. What is the difference between implicit and explicit type casting?**
- Implicit type casting happens automatically when converting data types, while explicit type casting requires a manual instuction from the programmer to convert a value from one type to another.

**11. What is the purpose of conditional statements in Python?**
- The purpose of conditional statement in Python is to control the flow of execuion in a program based on whether certain conditions are met. They enable the program to make dicisions and execute different blocks of code accordingly, the primary conditional statements in Python are "if", "elif", "else".

**12. How does the elif statement work?**
- The elif statement in Python is used in conditional statements to check multiple conditons one after another. It stands for "else if" and is used between "if" and "else".

**13. What is the difference between for and while loops?**
- Both for and while loops are used for repeating tasks, but they work differently.
  > **For loop:** Used when you knows how many times to repeat, it iterates over a sequence.

  > **while loop:** Used when you dont know the exact number of repetitions, it runs as long as a condition is True.

**14. Describe a scenario where a while loop is more suitable than a for loop.**
- A while loop is useful in an ATM system where a user has a limited number of attempts to enter the correct PIN. When a user inserts their card, the system prompts them to enter their PIN, if the PIN is correct, access is granted, adn the loop stops. However, if the user enters the wrong PIN, they are given another chance, with the number of attempt decreasing each time. If they fail to enter the correct PIN within the allowed attempts, the card is blocked. This scenario is ideal for a while loop  because the number of attempts needed is unknown in advance, and the loop runs until either the correct PIN is entered or the attempts run out.

**6. Write a function to demonstrate the behavior of mutable and immutable arguments.**

```python
def modify_values(num, my_list):
    num += 10  # This won't affect the original value outside the function (immutable)
    my_list.append(4)  # This modifies the original list (mutable)

# Immutable argument (integer)
x = 5
# Mutable argument (list)
lst = [1, 2, 3]

print("Before function call:")
print("x =", x)  # 5
print("lst =", lst)  # [1, 2, 3]

modify_values(x, lst)

print("\nAfter function call:")
print("x =", x)  # Still 5 (unchanged, because integers are immutable)
print("lst =", lst)  # Changed to [1, 2, 3, 4] (lists are mutable)
```

**14. Write a program to reverse a string using a while loop.**

```python
my_name = "Mashhood"
reversed_name = ''
index = len(my_name) -1

while index >= 0:
  reversed_name += my_name[index]
  index -= 1

print(reversed_name)

```

**15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.**

```python
user_num = int(input("Enter a number: "))
factorial = 1
counter = user_num

while counter > 0:
  factorial *= counter
  counter -= 1

print(factorial)

```
